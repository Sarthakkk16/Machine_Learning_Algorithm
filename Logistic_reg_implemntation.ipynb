```python
df= pd.DataFrame(data.data,columns=data.feature_names)
```

```python
df['target']=data.target
```

```python
# for binary classi

df = df[df['target']!= 2]
```

```python
df.target.unique()
```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
# model training

from sklearn.linear_model import LogisticRegression
classi = LogisticRegression()
```

```python
classi
```

```python
classi.fit(x_train,y_train)
```

```python
classi
```

```python
classi.coef_
```

```python
classi.intercept_
```

```python
y_pred = classi.predict(x_test)
```

```python
y_pred
```

```python
classi.predict_proba(x_test) # corresponding to higher prob class is predicted, cutoff is 0.5
```

```python
# evaluation metrics
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
confusion_matrix(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
# first two rows 0,1 >> precision recall and f1_score and number of data points wrt repected class.


# macroavg >> for wach class take simple avg >> avg precision of class 0 and class 1 will be macro avg similarly for recall, f1 score
# support will be total no of dps,

# weighted avg is used in case of class imbalance, it gives importance to performance of dp class.
```

```python
from sklearn.metrics import roc_curve, auc
classi.predict_proba(x_test)[:,-1]
```

```python
y_pred_proba = classi.predict_proba(x_test)[:,-1]
```

```python
y_pred_proba
```

```python
# ROC CURVE

fpr, tpr, thresholds = roc_curve(y_test,y_pred_proba)
```

```python
thresholds
```

```python
tpr
```

```python
fpr
```

```python
roc_auc=auc(fpr,tpr)
```

```python
#plot roc-auc curve

plt.figure(figsize=(8, 6))  # Create a new figure with a specified size
plt.plot(fpr, tpr, color='darkorange', linewidth=2, label='ROC curve (area = %0.2f)' % roc_auc)  # Plot ROC curve,
plt.plot([0, 1], [0, 1], color='navy', linewidth=2, linestyle='--')  # Plot the diagonal line representing random guessing
plt.xlim([0.0, 1.0])  # Set x-axis limits
plt.ylim([0.0, 1.05])  # Set y-axis limits
plt.xlabel('False Positive Rate')  # Set x-axis label
plt.ylabel('True Positive Rate')  # Set y-axis label
plt.title('Receiver Operating Characteristic (ROC) Curve')  # Set plot title
plt.legend(loc="lower right")  # Add legend to the plot
plt.show()  # Show the plot
```
