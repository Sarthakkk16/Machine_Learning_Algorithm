```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris
```

```python
data=load_iris()
```

```python
print(data.DESCR)
```

```python
df=pd.DataFrame(data.data,columns=data.feature_names)
```

```python
df
```

```python
data.target
```

```python
x=df
y=data.target
```

```python
x
```

```python
y
```

```python
# Train test split

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=10)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
classi=DecisionTreeClassifier(criterion='entropy')
classi
```

```python
classi.fit(x_train,y_train)
```

```python
y_pred=classi.predict(x_test)
```

```python
from sklearn import tree
plt.figure(figsize=(12,10))
tree.plot_tree(classi,filled=True)
```

```python
from sklearn import tree
plt.figure(figsize=(12,10))
tree.plot_tree(classi,filled=True)
plt.show() # if you dont want to see the details
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
# POST PRUNING>> after making complete dt you have to cut it to reduce overfitting, for datasets.

# observation >> observe the dt plot you will realise the last split is actually overfitting in such a way that each node is 100% pure leading to low no of data. So reduce the depth of the decision tree.
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier=DecisionTreeClassifier(criterion='entropy',max_depth=3)
classifier.fit(x_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(12,10))
tree.plot_tree(classifier,filled=True)
plt.show() # if you dont want to see the details
```

```python
y_pred=classifier.predict(x_test)
```

```python
accuracy_score(y_test,y_pred) # earlier it was a overfitting model.
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
# PRE PRUNING USING HYPERPARAMETER TUNING

parameter = {
    'criterion':['gini','entropy','log_loss'],
    'splitter':['best','random'],
    'max_depth':[2,3,4,5,6,7,8,9,10],
    'max_features':['auto','sqrt','log2']
}
```

```python
from sklearn.model_selection import GridSearchCV
clf=DecisionTreeClassifier()
model=GridSearchCV(clf,param_grid=parameter,cv=5,scoring='accuracy',verbose=3)
model.fit(x_train,y_train)
```

```python
model.best_params_
```

```python
model.best_score_
```

```python
y_pred=model.best_estimator_.predict(x_test)
```

```python
y_pred
```

```python
accuracy_score(y_test,y_pred)
```
